Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you already have:
# 1. A DataFrame called 'customers' with your customer data
# 2. Your data has been scaled and stored in a variable called 'X_scaled'

# Step 1: Initialize the K-Means model
# Let's use 4 clusters (you can adjust this number based on your specific requirements)
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)

# Step 2: Fit the model to your scaled data
kmeans_model = kmeans.fit(X_scaled)

# Step 3: Get the cluster labels
cluster_labels = kmeans_model.labels_

# Step 4: Add these labels back to your customers DataFrame
df['labels'] = cluster_labels

# Step 5: Now you can analyze the clusters
print(df_cleaned.groupby('labels').mean())

# Step 6: Visualize the clusters (if you have 2D data or use PCA to reduce dimensions)
# If you have already performed PCA and have components:
plt.figure(figsize=(10, 8))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=df['labels'], cmap='viridis', s=50, alpha=0.8)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, alpha=0.8, marker='X')
plt.title('K-Means Clustering Results')
plt.colorbar(label='Cluster')
plt.xlabel('Feature 1 (or PC1)')
plt.ylabel('Feature 2 (or PC2)')
plt.show()
```
